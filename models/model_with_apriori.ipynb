```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
import itertools 

from tkinter import ttk
```

```python
data_set=pd.read_excel(r'D:\MarketBasketAnalysis\datasets\retail_data.xlsx')
data_set
data_set.info()
```

```python
data_set['Itemname']=data_set['Itemname'].str.strip()
data_set.dropna(axis=0,subset=['BillNo'],inplace=True)
data_set['BillNo'] = data_set['BillNo'].astype('str')
data_set = data_set[~data_set['BillNo'].str.contains('C')]
data_set
```

```python
data_set['Country'].value_counts()
```

```python
one_basket = (data_set[data_set['Country'] =="United Kingdom"]
              .groupby(['BillNo','Itemname'])['Quantity']
              .sum().unstack().reset_index().fillna(0)
              .set_index(['BillNo']))
one_basket             
```

```python
def value_converter(value):
    if value>=1:
        return 1
    else:
         return 0
one_basket_sets = one_basket.applymap(value_converter)
one_basket_sets.drop('POSTAGE',inplace=True,axis=1)
one_basket_sets
```

```python
frequent_itemlist=apriori(one_basket_sets.astype(bool),min_support=0.03,use_colnames=True)
print(frequent_itemlist)
rules=association_rules(frequent_itemlist,metric="lift",min_threshold=1)
rules
```

```python
rules = rules[(rules['support']>=.035)&(rules['confidence']>=0.1)]
print(rules)
```

```python
wanted = rules[['antecedents','consequents']]
total_list = wanted.values.tolist()
new_list=[]
for item in total_list:
   antecedent = list(item[0])[0] 
   consequent = list(item[1])[0]
   new_list.append([antecedent,consequent])

print(new_list)
```

```python
unique_items = []
for item in new_list:
    for ele in item:
        if ele not in unique_items:
             unique_items.append(ele)
            
for one in unique_items:
    print(one)
```

```python
result_dictionary = {}
for itr in range(0,len(unique_items)):
    sets = set()
    for items in new_list:
        if unique_items[itr] in items:
             sets.update(items)
    result_dictionary[unique_items[itr]] = sets

for one,two in result_dictionary.items():
    print(f"{one}: {two}")
```

```python
import tkinter as tk
from PIL import Image,ImageTk

class CombinationApp:
    def __init__(self, root):
        root.title("Market Basket Analysis")

        main_heading_label = ttk.Label(root, text="Market Basket Analysis", font=('Arial', 20, 'bold'))
        main_heading_label.pack(pady=15)

        background_image = Image.open("mba0.jpg")  
        background_photo = ImageTk.PhotoImage(background_image)

        self.unique_items = [
            #'ALARM CLOCK BAKELIKE RED',
            #'ALARM CLOCK BAKELIKE GREEN',
            'GREEN REGENCY TEACUP AND SAUCER',
            'ROSES REGENCY TEACUP AND SAUCER',
            'JUMBO BAG PINK POLKADOT',
            'JUMBO BAG RED RETROSPOT',
            'JUMBO SHOPPER VINTAGE RED PAISLEY',
            'JUMBO STORAGE BAG SUKI',
            #'LUNCH BAG RED RETROSPOT',
            #'LUNCH BAG BLACK SKUL'
        ]

        self.result_dictionary = {
            'GREEN REGENCY TEACUP AND SAUCER': {'GREEN REGENCY TEACUP AND SAUCER', 'ROSES REGENCY TEACUP AND SAUCER'},
            'ROSES REGENCY TEACUP AND SAUCER': {'GREEN REGENCY TEACUP AND SAUCER', 'ROSES REGENCY TEACUP AND SAUCER'},
            'JUMBO BAG PINK POLKADOT': {'JUMBO BAG PINK POLKADOT', 'JUMBO BAG RED RETROSPOT'},
            'JUMBO BAG RED RETROSPOT': {'JUMBO BAG PINK POLKADOT', 'JUMBO BAG RED RETROSPOT', 'JUMBO SHOPPER VINTAGE RED PAISLEY', 'JUMBO STORAGE BAG SUKI'},
            'JUMBO SHOPPER VINTAGE RED PAISLEY': {'JUMBO BAG RED RETROSPOT', 'JUMBO SHOPPER VINTAGE RED PAISLEY'},
            'JUMBO STORAGE BAG SUKI': {'JUMBO BAG RED RETROSPOT', 'JUMBO STORAGE BAG SUKI'}
        }

        self.combination_label = ttk.Label(root, text="Select an item:", font=('Times New Roman', 12))
        self.combination_label.pack(pady=10)

        self.combination_var = tk.StringVar()
        self.combination_dropdown = ttk.Combobox(root, textvariable=self.combination_var, values=self.unique_items, state="readonly", font=('Times New Roman', 16), height = 20)
        self.combination_dropdown.bind("<<ComboboxSelected>>", self.update_combinations)
        self.combination_dropdown.pack(pady=15)

        self.combination_table = ttk.Treeview(root, columns=('Combinations',), show='headings', height=20)
        self.combination_table.heading('Combinations', text='Combinations')
        self.combination_table.column('Combinations', width= 800 )
        self.combination_table.pack(pady=15)

        self.combination_table.tag_configure('mytag', font=('Times New Roman', 14, 'bold'))


    @staticmethod
    def combinator(item, result_dictionary):
        related_items = result_dictionary[item]
        possible_list = []
        for i in range(1, len(related_items) + 1):
            possible_list.extend(itertools.combinations(related_items, i))
        return possible_list

    def update_combinations(self, event):
        selected_item = self.combination_var.get()
        combinations = self.combinator(selected_item, self.result_dictionary)

        self.combination_table.delete(*self.combination_table.get_children())

        for combo in combinations:
            formatted_combo = ', '.join(combo)
            self.combination_table.insert('', 'end', values=(formatted_combo,))

if __name__ == "__main__":
    root = tk.Tk()
    root.geometry("1200x800")  # Adjust the initial size of the window
    app = CombinationApp(root)
    root.mainloop()
```
