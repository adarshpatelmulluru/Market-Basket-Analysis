```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data_set=pd.read_excel('retail_data.xlsx')
data_set
```

```python
data_set['Itemname']=data_set['Itemname'].str.strip()
data_set.dropna(axis=0,subset=['BillNo'],inplace=True)
data_set['BillNo'] = data_set['BillNo'].astype('str')
data_set = data_set[~data_set['BillNo'].str.contains('C')]
data_set
```

```python
one_basket = (data_set[data_set['Country'] =="United Kingdom"]
              .groupby(['BillNo','Itemname'])['Quantity']
              .sum().unstack().reset_index().fillna(0)
              .set_index(['BillNo']))
one_basket             
```

```python
def value_converter(value):
    if value>=1:
        return 1
    else:
         return 0
one_basket_sets = one_basket.applymap(value_converter)
one_basket_sets.drop('POSTAGE',inplace=True,axis=1)
one_basket_sets
```

```python
frequent_itemlist=apriori(one_basket_sets.astype(bool),min_support=0.03,use_colnames=True)
print(frequent_itemlist)
rules=association_rules(frequent_itemlist,metric="lift",min_threshold=1)
rules
```

```python
rules[(rules['lift']>=4)&(rules['confidence']>=0.5)]
```
